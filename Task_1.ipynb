```python
import altair as alt
from altair import datum
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('GLB.Ts_dSST.csv', skiprows = 1, usecols=range(13), index_col=0)
```

```python
df.head()
```

```python
df['y_avg'] = df.mean(axis=1)
xix_avg = df.loc[range(1880,1900)]['y_avg'].mean()
df.reset_index(inplace = True)
```

```python
df.head()
```

```python
line = alt.Chart(pd.DataFrame({'coord_y': [xix_avg]})).mark_rule().encode(y='coord_y')

chart = alt.Chart(
    df
    ).mark_point(
         color = '#545454'
        ,fill = '#F26C08'
        ,size = 50.26548245743669
        ,opacity = 1
        ,strokeWidth = 1
    ).encode(
         x = alt.X(
                     field = 'Year'
                    ,title = None
                    ,type = 'ordinal'
                    ,axis = alt.Axis(
                                 values = list(range(1880,2017,10))
                                ,domain = False
                                ,offset = 10
                                ,labelFont = 'nytfranklin'
                                ,labelFontWeight = 300
                                ,labelFontSize = 12
                                ,labelColor = '#999'
                            )
                   )
        ,y = alt.Y(
                     field = 'y_avg'
                    ,title = None
                    ,type = 'quantitative'
                    ,scale = alt.Scale(domain = [-0.8, 1.2])
                    ,axis = alt.Axis(
                                 offset = 8
                                ,values = np.round(np.arange(-0.6, 1.2, 0.2), 2)
                                ,labelExpr = "if(datum.value == 0, '' , if(datum.value > 0, '+', '') ) + datum.value + if((datum.value % 1) == 0 , '.0', '') + '\u00b0' + if(datum.index == 1, 'C', '') " 
                                ,labelFont = 'nytfranklin'
                                ,labelFontWeight = 300
                                ,labelFontSize = 12
                                ,labelColor = '#999'
                            )
                   )
    )

text_df = pd.DataFrame({'text_hot': ['HOTTER THAN THE \n1880-1889 AVERAGE']
                       ,'text_cold': ['COLDER']
                       ,'title' : ['Annual Global Surface Temperature, \nRelative to Late 19th Century Average']
                       ,'source' : ['Source: NASA']
                       ,'coord_y': [xix_avg]
                       ,'coord_x': [2010]})

hot_text = alt.Chart(text_df).mark_text(
         color = '#000'
        ,lineBreak='\n'
        ,fontSize = 11.2
        ,font = 'nytfranklin'
        ,fontWeight = 300
    ).encode(
        text = alt.Text(field = 'text_hot'
                       ,type = 'nominal'
                       )
       ,x = alt.value(890)
       ,y = alt.value(360)
    )

cold_text = alt.Chart(text_df).mark_text(
                 color = '#000'
                ,fontSize = 11.2
                ,font = 'nytfranklin'
                ,fontWeight = 300
            ).encode(
                text = alt.Text(field = 'text_cold'
                               ,type = 'nominal'
                               )     
           ,x = alt.value(920)
           ,y = alt.value(410)
        )

title_text = alt.Chart(text_df).mark_text(
                  color = '#333'
                 ,fontSize = 16
                 ,font = 'nytfranklin'
                 ,fontWeight = 800
                 ,lineBreak='\n'
                 ,align='left'
            ).encode(
                text = alt.Text(field = 'title'
                               ,type = 'nominal'
                               )
              ,x = alt.value(25)
              ,y = alt.value(55)
            )

source_text = alt.Chart(text_df).mark_text(
                   color = '#999'
                  ,fontSize = 11
                  ,font = 'nytfranklin'
                  ,fontWeight = 400
            ).encode(
                text = alt.Text(field = 'source'
                               ,type = 'nominal'
                               )
                ,x = alt.value(-11)
                ,y = alt.value(600)
            )

labels_left = chart.mark_text(
                color = '#333'
               ,fontSize = 12
               ,font = 'nytfranklin'
               ,fontWeight = 500
               ,dy = 0
               ,dx = -22
        ).encode(
              text = alt.Text(field = 'Year'
                             ,type = 'nominal')
        ).transform_filter(
            alt.FieldOneOfPredicate(field='Year', oneOf = [1904, 1998])
        )

labels_right = chart.mark_text(
                color = '#333'
               ,fontSize = 12
               ,font = 'nytfranklin'
               ,fontWeight = 500
               ,dy = 0
               ,dx = 22
        ).encode(
              text = alt.Text(field = 'Year'
                             ,type = 'nominal')
        ).transform_filter(
            alt.FieldOneOfPredicate(field='Year', oneOf = [1944, 2014, 2017])
        )

labels_2015 = chart.mark_text(
                color = '#333'
               ,fontSize = 12
               ,font = 'nytfranklin'
               ,fontWeight = 500
               ,dy = 3
               ,dx = 22
        ).encode(
              text = alt.Text(field = 'Year'
                             ,type = 'nominal')
        ).transform_filter(
            alt.FieldOneOfPredicate(field='Year', oneOf = [2015])
        )

labels_2016 = chart.mark_text(
                color = '#333'
               ,fontSize = 12
               ,font = 'nytfranklin'
               ,fontWeight = 600
#                ,dy = 1
               ,dx = 22
        ).encode(
              text = alt.Text(field = 'Year'
                             ,type = 'nominal')
        ).transform_filter(
            alt.FieldOneOfPredicate(field='Year', oneOf = [2016])
        )

alt.layer(line, chart, hot_text, cold_text, title_text, source_text, labels_left, labels_right, labels_2015, labels_2016
    ).properties(
         width = 945
        ,height = 550
        ,background = 'white'
        ,padding = 25
    ).configure_mark(
        strokeWidth = 0.5
    ).configure_axis(
         ticks = False
        ,grid = True
        ,gridColor = '#e1e1e1'
        ,gridOpacity = 0.5
        ,labelAngle = 0
        ,domain = False
    ).configure_view(
        strokeWidth = 0
    )
```
